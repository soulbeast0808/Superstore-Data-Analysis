```python
# EXPLORARTORY DATA ANALYSIS (EDA)
#total sales 

total_sales = df["Sales"].sum()
print(total_sales)
```

```python
#TOTAL PROFIT 
total_profit = df["Profit"].sum()
print(total_profit)
```

```python
#sales by category 
category_sales = df.groupby("Category")["Sales"].sum()
print(category_sales)
```

```python
# visualization 
import matplotlib.pyplot as plt
category_sales.plot(kind="bar")
plt.title("Sales by Category")
plt.show()
```

```python
#SALES BY REGION
import matplotlib.pyplot as plt
region_sales = df.groupby("Region")["Sales"].sum()
region_sales.plot(kind="bar")
plt.title("Sales by Region")
plt.show()
```

```python
#TOP 10 PRODUCT FOR SALES 
import seaborn as sns
top_products = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)

top_products.plot(kind="bar")
plt.title("Top 10 Products by Sales")
plt.show()
```

```python
#TOP 5 PRODUCT BY SALES 
import seaborn as sns
top_products = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(5)

top_products.plot(kind="bar")
plt.title("Top 5 Products by Sales")
plt.show()
```

```python
#TOP 3 PRODUCT BY SALES 
import seaborn as sns
top_products = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(3)

top_products.plot(kind="bar")
plt.title("Top 3 Products by Sales")
plt.show()
```

```python
#MONTHLY SALES TRENDS 
import numpy as np
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
import numpy as np
df["Month"] = df["Order Date"].dt.month
```

```python
import numpy as np
monthly_sales = df.groupby("Month")["Sales"].sum()

monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.show()
```

```python
#CREATE CORRELATION HEATMAP

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# load dataset
df = pd.read_csv("OneDrive/sample super store.csv", encoding="latin1")
# correlation heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```
